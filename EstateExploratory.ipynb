```python
import pandas as pd
import matplotlib.pyplot as plt
```

# --- 1. Load the cleaned data ---

```python
df = pd.read_csv('./Data/cleaned_housePrice.csv')
df
```

# --- 2. Visualize the distribution of 'Price' ---

```python
# A histogram helps visualize the frequency distribution of prices.
print("\n--- Generating Price Distribution Histogram ---")
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Property Prices', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
#plt.savefig('price_distribution.png')
print("Histogram saved as 'price_distribution.png'.")
```

# --- 3. Visualize the relationship between 'Area' and 'Price' ---

```python
# A scatter plot is great for showing the correlation between two numerical variables.
print("\n--- Generating Area vs. Price Scatter Plot ---")
plt.figure(figsize=(10, 6))
plt.scatter(df['Area'], df['Price'], alpha=0.5, color='coral')
plt.title('Area vs. Price', fontsize=16)
plt.xlabel('Area', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(True)
#plt.savefig('area_vs_price.png')
print("Scatter plot saved as 'area_vs_price.png'.")
```

# --- 4. Analyze the impact of categorical features on 'Price' ---

# A bar chart is useful for comparing the mean price across different categories.

```python
# Impact of 'Elevator' on Price
print("\n--- Generating Elevator Impact on Price Bar Chart ---")
if 'Elevator' in df.columns:
    elevator_price = df.groupby('Elevator')['Price'].mean().sort_values(ascending=False)
    elevator_price.plot(kind='bar', color=['lightgreen', 'salmon'])
    plt.title('Average Price by Elevator Presence', fontsize=16)
    plt.xlabel('Elevator (1 = Yes, 0 = No)', fontsize=12)
    plt.ylabel('Average Price', fontsize=12)
    plt.xticks(rotation=0)
    #plt.savefig('elevator_impact.png')
    print("Bar chart saved as 'elevator_impact.png'.")
else:
    print("Warning: 'Elevator' column not found. Skipping plot.")
```

```python
# Impact of 'Parking' on Price
print("\n--- Generating Parking Impact on Price Bar Chart ---")
if 'Parking' in df.columns:
    parking_price = df.groupby('Parking')['Price'].mean().sort_values(ascending=False)
    parking_price.plot(kind='bar', color=['darkblue', 'orange'])
    plt.title('Average Price by Parking Presence', fontsize=16)
    plt.xlabel('Parking (1 = Yes, 0 = No)', fontsize=12)
    plt.ylabel('Average Price', fontsize=12)
    plt.xticks(rotation=0)
    #plt.savefig('parking_impact.png')
    print("Bar chart saved as 'parking_impact.png'.")
else:
    print("Warning: 'Parking' column not found. Skipping plot.")
```
