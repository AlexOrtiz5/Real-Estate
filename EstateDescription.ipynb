```python
import pandas as pd
```

# --- 1. Load the cleaned data ---

```python
df = pd.read_csv('./Data/cleaned_housePrice.csv')
df
```

# --- 2. Descriptive Statistics for Numerical Columns ---

```python
# This will show key stats like count, mean, std, min, max, and quartiles.
print("\n--- Descriptive Statistics for Numerical Columns ---")
numerical_columns = ['Area', 'Room', 'Price', 'Price(USD)']
print(df[numerical_columns].describe())
```

# --- 3. Property Counts for Categorical Features ---

```python
# This section will count how many properties have each of the specified amenities.
print("\n--- Property Counts for Categorical Features ---")

# 'Parking' count
if 'Parking' in df.columns:
    parking_counts = df['Parking'].value_counts()
    print("\nParking counts:")
    print(parking_counts)
else:
    print("Warning: 'Parking' column not found.")

# 'Warehouse' count
if 'Warehouse' in df.columns:
    warehouse_counts = df['Warehouse'].value_counts()
    print("\nWarehouse counts:")
    print(warehouse_counts)
else:
    print("Warning: 'Warehouse' column not found.")

# 'Elevator' count
if 'Elevator' in df.columns:
    elevator_counts = df['Elevator'].value_counts()
    print("\nElevator counts:")
    print(elevator_counts)
else:
    print("Warning: 'Elevator' column not found.")

# You can also use a similar approach for 'Room'
print("\nRoom counts:")
if 'Room' in df.columns:
    room_counts = df['Room'].value_counts().sort_index()
    print(room_counts)
else:
    print("Warning: 'Room' column not found.")
```
